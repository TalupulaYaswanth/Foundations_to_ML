```python
data={
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'age': [25, 30, 35, 40, 45],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}#    |---->this name is the column name, and this list is the values of the column, and the length of the list should be the same for all columns, and the number of columns should be the same for all rows, and the number of rows should be the same for all columns, and so on.
df=pb.DataFrame(data,index=['stu1', 'stu2', 'stu3', 'stu4', 'stu5'])# we can create a dataframe from a dictionary, and the keys of the dictionary will be the column names of the dataframe, and the values of the dictionary will be the values of the columns of the dataframe, and the index of the dataframe will be the index of the dictionary, and so on.   
print(df)
df.set_index('name', inplace=True)# we can set the index of the dataframe to a column, and the column will be removed from the dataframe, and the index will be set to the values of the column, and so on.
print(df)
df.set_flags(allows_duplicate_labels=False)# we can allow duplicate labels in the index of the dataframe, and the duplicate labels will be allowed in the index, and so on.
print(df)
```

```python
df3.dropna()   # dropna() method is used to remove the rows that contain missing values from the dataframe df3. It will return a new dataframe that contains only the rows that do not have any missing values. By default, it will remove any row that has at least one missing value, but you can also specify how to handle missing values using the parameters of the dropna() method.
```
